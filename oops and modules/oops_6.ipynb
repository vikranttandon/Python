# MRO (method resolution order)

This is a __diamond case__ 

- Here, if (class D -> class E) means that class D inherits from class E,
  If _class E_ has an attribute 'x' and _class B_ also has an attribute 'x', which 'x' will class D       inherit first?

Python solves this problem with the help of __C3 linearization__ algorithm,
- This algo works much like a DFS algo,
- Starting the traversal from the class (here D) whose object has been created,
  it further visits other nodes (classes) only if they satisfy a certain condition.
  - condition:
    - Only to visit a particular node (class) if all of its children nodes (classes) have been visited.
      
- Here, there are multiple orders possible (even using c3 linearization algo) but in all the cases,
  the attribute 'x' of _class E_ will be inherited by _class D_ earlier than that of _class B_,
  because after visiting _class C_, when it'll try to visit _class B_ it will notice that the _class E_
  (which is a child class of _class B_) has not yet been visited, hence it will visit _class E_ first.

```python
class A:
    pass
class B:
    x = 'class B'
class E(A,B):
    x = 'class E'
class C(B):
    pass
class D(C,E):
    def show(self):
        print(self.x)
```

```python
aux = D()
```

```python
aux.show()
```

We can check out the order of method resolution :

```python
D.__mro__
```
